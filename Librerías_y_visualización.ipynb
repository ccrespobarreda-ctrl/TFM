# 2. Exploraci√≥n b√°sica

```python
# =============================================================================
# AN√ÅLISIS EXPLORATORIO Y PREPROCESAMIENTO - DATASET BRFSS
# =============================================================================

# ------------------------------------
# 1. LIBRER√çAS Y CONFIGURACI√ìN
# ------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
import warnings
warnings.filterwarnings('ignore')

# Configuraci√≥n de visualizaci√≥n
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)

print("‚úÖ Librer√≠as cargadas y configuradas")

# ------------------------------------
# 2. CARGA OPTIMIZADA DEL DATASET
# ------------------------------------
def cargar_dataset_optimizado(ruta):
    """
    Carga el dataset optimizando tipos de datos para reducir memoria
    """
    print("üì• Cargando dataset con optimizaci√≥n de memoria...")

    # Primera pasada para identificar tipos de datos
    # Read a sample to infer dtypes, but be cautious with floats
    sample_dataset = pd.read_csv(ruta, encoding='Latin-1', nrows=1000)

    # Define optimized types based on initial sample
    dtype_optimizado = {}
    for col in sample_dataset.columns:
        if sample_dataset[col].dtype == 'object':
            # For object columns, check if they can be categorical
            unique_ratio = sample_dataset[col].nunique() / len(sample_dataset)
            if unique_ratio < 0.5:
                dtype_optimizado[col] = 'category'
        elif np.issubdtype(sample_dataset[col].dtype, np.integer):
            # Optimize integer types
            col_min = sample_dataset[col].min()
            col_max = sample_dataset[col].max()

            if col_min >= 0:
                if col_max < 256:
                    dtype_optimizado[col] = 'uint8'
                elif col_max < 65536:
                    dtype_optimizado[col] = 'uint16'
                elif col_max < 4294967295:
                    dtype_optimizado[col] = 'uint32'
                else:
                    dtype_optimizado[col] = 'uint64'
            else:
                if col_min > -128 and col_max < 127:
                    dtype_optimizado[col] = 'int8'
                elif col_min > -32768 and col_max < 32767:
                    dtype_optimizado[col] = 'int16'
                elif col_min > -2147483648 and col_max < 2147483647:
                    dtype_optimizado[col] = 'int32'
                else:
                    dtype_optimizado[col] = 'int64'
        # For float types, let pandas infer or keep float64 for safety
        elif np.issubdtype(sample_dataset[col].dtype, np.floating):
             # Keep float64 or let pandas infer if potential for large/negative numbers
             dtype_optimizado[col] = 'float64' # Be explicit for safety


    # Full load with optimized types
    dataset_completo = pd.read_csv(ruta, encoding='Latin-1', dtype=dtype_optimizado, low_memory=False)

    memoria_mb = dataset_completo.memory_usage(deep=True).sum() / 1024**2
    print(f"‚úÖ Dataset cargado: {dataset_completo.shape[0]:,} filas √ó {dataset_completo.shape[1]} columnas")
    print(f"üíæ Memoria utilizada: {memoria_mb:.2f} MB")

    return dataset_completo

# Mount Drive and load data
drive.mount('/content/drive')
ruta = '/content/drive/MyDrive/Copia de Dataset_sin_tratar.csv'
dataset_original = cargar_dataset_optimizado(ruta)

# ------------------------------------
# 3. FULL EXPLORATORY ANALYSIS
# ------------------------------------
def analisis_exploratorio_completo(dataset):
    """
    Exhaustive analysis of the dataset following best practices
    """
    print("\n" + "="*80)
    print("üîç FULL EXPLORATORY ANALYSIS")
    print("="*80)

    # Basic Information
    print("\nüìä GENERAL INFORMATION:")
    print(f"‚Ä¢ Dimensions: {dataset.shape[0]:,} rows √ó {dataset.shape[1]} columns")
    print(f"‚Ä¢ Total memory: {dataset.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

    # Data Types Distribution
    print("\nüìà DATA TYPES DISTRIBUTION:")
    tipo_counts = dataset.dtypes.value_counts()
    for tipo, count in tipo_counts.items():
        print(f"‚Ä¢ {tipo}: {count} columns ({count/dataset.shape[1]*100:.1f}%)")

    # Quick Descriptive Statistics
    print("\nüìã DESCRIPTIVE STATISTICS (first 10 numeric columns):")
    numeric_cols = dataset.select_dtypes(include=np.number).columns[:10]
    display(dataset[numeric_cols].describe())

    return dataset

dataset_original = analisis_exploratorio_completo(dataset_original)

# ------------------------------------
# 4. MISSING VALUES ANALYSIS
# ------------------------------------
def analisis_missing_values(dataset, umbral_alto=70, umbral_medio=30):
    """
    Detailed analysis of missing values with treatment strategies
    """
    print("\n" + "="*80)
    print("üîç MISSING VALUES ANALYSIS")
    print("="*80)

    total_filas = len(dataset)
    nulos_por_columna = dataset.isnull().sum()
    porcentaje_nulos = (nulos_por_columna / total_filas) * 100

    resumen_nulos = pd.DataFrame({
        'valores_nulos': nulos_por_columna,
        'porcentaje_nulos': porcentaje_nulos,
        'tipo_dato': dataset.dtypes
    }).sort_values('porcentaje_nulos', ascending=False)

    # Classify columns by missing value level
    columnas_problematicas = resumen_nulos[resumen_nulos['porcentaje_nulos'] > umbral_alto]
    columnas_manejables = resumen_nulos[
        (resumen_nulos['porcentaje_nulos'] > umbral_medio) &
        (resumen_nulos['porcentaje_nulos'] <= umbral_alto)
    ]
    columnas_limpias = resumen_nulos[resumen_nulos['porcentaje_nulos'] <= umbral_medio]

    print(f"\nüìä MISSING VALUES SUMMARY:")
    print(f"‚Ä¢ Columns with >{umbral_alto}% nulls: {len(columnas_problematicas)}")
    print(f"‚Ä¢ Columns with {umbral_medio}-{umbral_alto}% nulls: {len(columnas_manejables)}")
    print(f"‚Ä¢ Columns with <{umbral_medio}% nulls: {len(columnas_limpias)}")
    print(f"‚Ä¢ Total null values in dataset: {dataset.isnull().sum().sum():,}")

    # Show the top 10 most problematic columns
    if len(columnas_problematicas) > 0:
        print(f"\nüö® TOP 10 MOST PROBLEMATIC COLUMNS (> {umbral_alto}% nulls):")
        display(columnas_problematicas.head(10))

    return resumen_nulos, columnas_problematicas, columnas_manejables, columnas_limpias

resumen_nulos, problematicas, manejables, limpias = analisis_missing_values(dataset_original)

# ------------------------------------
# 5. TARGET VARIABLE ANALYSIS (DIABETE3)
# ------------------------------------
def analisis_variable_target(dataset, target_col='DIABETE3'):
    """
    Exhaustive analysis of the target variable for classification problems
    """
    print("\n" + "="*80)
    print("üéØ TARGET VARIABLE ANALYSIS: DIABETE3")
    print("="*80)

    # BRFSS code dictionary (according to official documentation)
    codigos_diabetes = {
        1: "S√≠, diabetes",
        2: "S√≠, but only during pregnancy",
        3: "No, does not have diabetes",
        4: "S√≠, prediabetes or borderline",
        7: "Does not know / Not sure",
        9: "Refused to answer"
    }

    # Detailed distribution
    conteos = dataset[target_col].value_counts().sort_index()
    porcentajes = (conteos / len(dataset)) * 100

    print("\nüìä TARGET VARIABLE DISTRIBUTION:")
    for valor, conteo in conteos.items():
        etiqueta = codigos_diabetes.get(valor, f"Valor {valor}")
        print(f"  {valor}: {etiqueta:.<40} {conteo:>8,} ({porcentajes[valor]:.2f}%)")

    # Analysis of useful classes for modeling
    clases_principales = {1: "Diabetes", 3: "No Diabetes", 4: "Prediabetes"}
    clases_excluir = {2: "Gestational", 7: "Does not know", 9: "Refused"}

    total_util = sum(conteos.get(k, 0) for k in clases_principales.keys())
    total_excluir = sum(conteos.get(k, 0) for k in clases_excluir.keys())

    print(f"\nüéØ CLASSES FOR MODELING:")
    for valor, etiqueta in clases_principales.items():
        if valor in conteos:
            count = conteos[valor]
            print(f"  ‚Ä¢ {etiqueta}: {count:,} records ({count/len(dataset)*100:.2f}%)")

    print(f"\nüö´ CLASSES TO EXCLUDE:")
    for valor, etiqueta in clases_excluir.items():
        if valor in conteos:
            count = conteos[valor]
            print(f"  ‚Ä¢ {etiqueta}: {count:,} records ({count/len(dataset)*100:.2f}%)")

    print(f"\nüìà CLASS BALANCE:")
    print(f"  ‚Ä¢ Useful records for modeling: {total_util:,} ({total_util/len(dataset)*100:.2f}%)")
    print(f"  ‚Ä¢ Records to exclude: {total_excluir:,} ({total_excluir/len(dataset)*100:.2f}%)")

    if total_util > 0:
        for valor, etiqueta in clases_principales.items():
            if valor in conteos:
                proporcion = conteos[valor] / total_util * 100
                print(f"  ‚Ä¢ {etiqueta} in useful dataset: {conteos[valor]:,} ({proporcion:.1f}%)")

    # Visualization of the distribution
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    datos_plot = dataset[target_col].value_counts().sort_index()
    bars = plt.bar([codigos_diabetes.get(x, x) for x in datos_plot.index], datos_plot.values)
    plt.title('Distribution of DIABETE3', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Frequency')

    # Add labels to bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:,}\n({height/len(dataset)*100:.1f}%)',
                ha='center', va='bottom', fontsize=8)

    plt.subplot(1, 2, 2)
    # Only main classes for pie chart
    clases_para_pie = {k: v for k, v in clases_principales.items() if k in conteos}
    sizes = [conteos[k] for k in clases_para_pie.keys()]
    labels = [f'{v}\n({sizes[i]/total_util*100:.1f}%)' for i, v in enumerate(clases_para_pie.values())]

    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')
    plt.title('Distribution of Main Classes\n(for modeling)', fontsize=14, fontweight='bold')

    plt.tight_layout()
    plt.show()

    return clases_principales, total_util

clases_principales, total_util = analisis_variable_target(dataset_original)

# ------------------------------------
# 6. CARDINALITY AND CONSTANT VARIABLES ANALYSIS
# ------------------------------------
def analisis_cardinalidad_constantes(dataset):
    """
    Identifies variables with high cardinality and constant variables
    """
    print("\n" + "="*80)
    print("üìä CARDINALITY AND CONSTANT VARIABLES ANALYSIS")
    print("="*80)

    total_filas = len(dataset)
    resumen_cardinalidad = []

    for columna in dataset.columns:
        valores_unicos = dataset[columna].nunique()
        porcentaje_unicos = (valores_unicos / total_filas) * 100
        es_constante = valores_unicos == 1
        tipo_dato = dataset[columna].dtype

        resumen_cardinalidad.append({
            'columna': columna,
            'valores_unicos': valores_unicos,
            'porcentaje_unicos': porcentaje_unicos,
            'es_constante': es_constante,
            'tipo_dato': tipo_dato
        })

    df_cardinalidad = pd.DataFrame(resumen_cardinalidad)

    # Constant variables
    variables_constantes = df_cardinalidad[df_cardinalidad['es_constante']]
    print(f"\nüö® CONSTANT VARIABLES (recommended to drop): {len(variables_constantes)}")
    if len(variables_constantes) > 0:
        display(variables_constantes[['columna', 'valores_unicos', 'tipo_dato']])

    # Variables with high cardinality (>90% unique values)
    alta_cardinalidad = df_cardinalidad[df_cardinalidad['porcentaje_unicos'] > 90]
    print(f"\n‚ö†Ô∏è VARIABLES WITH HIGH CARDINALITY (>90% unique): {len(alta_cardinalidad)}")
    if len(alta_cardinalidad) > 0:
        display(alta_cardinalidad[['columna', 'valores_unicos', 'porcentaje_unicos', 'tipo_dato']].head(10))

    # Variables with low cardinality (<5% unique values - potential categorical)
    baja_cardinalidad = df_cardinalidad[
        (df_cardinalidad['porcentaje_unicos'] < 5) &
        (~df_cardinalidad['es_constante']) &
        (df_cardinalidad['valores_unicos'] > 1)
    ]
    print(f"\nüéØ VARIABLES WITH LOW CARDINALITY (<5% unique): {len(baja_cardinalidad)}")
    if len(baja_cardinalidad) > 0:
        display(baja_cardinalidad[['columna', 'valores_unicos', 'porcentaje_unicos', 'tipo_dato']].head(10))

    return df_cardinalidad, variables_constantes, alta_cardinalidad, baja_cardinalidad

df_cardinalidad, constantes, alta_card, baja_card = analisis_cardinalidad_constantes(dataset_original)

# ------------------------------------
# 7. CREATING THE CLEAN DATASET
# ------------------------------------
def crear_dataset_limpio(dataset, target_col='DIABETE3', umbral_missing=70):
    """
    Creates a clean version of the dataset applying quality criteria
    """
    print("\n" + "="*80)
    print("üßπ CREATING CLEAN DATASET")
    print("="*80)

    dataset_limpio = dataset.copy()
    cambios = []

    # Step 1: Filter only useful records from the target variable
    registros_originales = len(dataset_limpio)
    dataset_limpio = dataset_limpio[dataset_limpio[target_col].isin([1, 3, 4])]
    registros_filtrados = len(dataset_limpio)
    cambios.append(f"Filtered target: {registros_originales:,} ‚Üí {registros_filtrados:,} records")

    # Step 2: Remove columns with many missing values
    columnas_originales = dataset_limpio.shape[1]
    porcentaje_nulos = dataset_limpio.isnull().mean() * 100
    columnas_a_eliminar = porcentaje_nulos[porcentaje_nulos > umbral_missing].index.tolist()

    dataset_limpio = dataset_limpio.drop(columns=columnas_a_eliminar)
    columnas_finales = dataset_limpio.shape[1]
    cambios.append(f"Removed columns: {columnas_originales} ‚Üí {columnas_finales} ({len(columnas_a_eliminar)} removed)")

    # Step 3: Remove constant variables
    variables_constantes = [col for col in dataset_limpio.columns if dataset_limpio[col].nunique() == 1]
    dataset_limpio = dataset_limpio.drop(columns=variables_constantes)
    cambios.append(f"Removed constants: {len(variables_constantes)} columns")

    # Step 4: Remove columns with high cardinality (possible IDs)
    total_filas = len(dataset_limpio)
    columnas_alta_cardinalidad = []
    for col in dataset_limpio.columns:
        if dataset_limpio[col].nunique() / total_filas > 0.9 and col != target_col:
            columnas_alta_cardinalidad.append(col)

    dataset_limpio = dataset_limpio.drop(columns=columnas_alta_cardinalidad)
    cambios.append(f"Removed high cardinality: {len(columnas_alta_cardinalidad)} columns")

    # Final Summary
    print("üìã CLEANUP SUMMARY:")
    for cambio in cambios:
        print(f"  ‚Ä¢ {cambio}")

    print(f"\n‚úÖ CLEAN DATASET CREATED:")
    print(f"   ‚Ä¢ Dimensions: {dataset_limpio.shape[0]:,} rows √ó {dataset_limpio.shape[1]} columns")
    print(f"   ‚Ä¢ Memory: {dataset_limpio.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    print(f"   ‚Ä¢ Columns removed: {dataset.shape[1] - dataset_limpio.shape[1]}")

    if len(columnas_a_eliminar) > 0:
        print(f"\nüö´ COLUMNS REMOVED DUE TO HIGH % OF MISSING VALUES:")
        for col in columnas_a_eliminar[:10]:  # Show only the first 10
            print(f"   ‚Ä¢ {col} ({porcentaje_nulos[col]:.1f}% nulls)")
        if len(columnas_a_eliminar) > 10:
            print(f"   ‚Ä¢ ... and {len(columnas_a_eliminar) - 10} more")

    return dataset_limpio, columnas_a_eliminar

dataset_limpio, columnas_eliminadas = crear_dataset_limpio(dataset_original)

# ------------------------------------
# 8. FINAL ANALYSIS OF CLEAN DATASET
# ------------------------------------
def analisis_final_dataset_limpio(dataset, target_col='DIABETE3'):
    """
    Final analysis of the clean dataset to verify quality
    """
    print("\n" + "="*80)
    print("‚úÖ FINAL ANALYSIS - CLEAN DATASET")
    print("="*80)

    # General Information
    print("üìä GENERAL INFORMATION:")
    print(f"  ‚Ä¢ Dimensions: {dataset.shape[0]:,} rows √ó {dataset.shape[1]} columns")
    print(f"  ‚Ä¢ Memory used: {dataset.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

    # Data Types Distribution
    print(f"\nüìà DATA TYPES:")
    for tipo, count in dataset.dtypes.value_counts().items():
        print(f"  ‚Ä¢ {tipo}: {count} columns")

    # Missing values in clean dataset
    nulos_totales = dataset.isnull().sum().sum()
    nulos_por_columna = dataset.isnull().mean() * 100
    columnas_con_nulos = nulos_por_columna[nulos_por_columna > 0]

    print(f"\nüîç MISSING VALUES:")
    print(f"  ‚Ä¢ Total null values: {nulos_totales:,}")
    print(f"  ‚Ä¢ Columns with null values: {len(columnas_con_nulos)}")
    print(f"  ‚Ä¢ Total percentage of nulls: {nulos_totales / (dataset.shape[0] * dataset.shape[1]) * 100:.2f}%")

    if len(columnas_con_nulos) > 0:
        print(f"\nüìã TOP 10 COLUMNS WITH MOST MISSING VALUES:")
        display(pd.DataFrame({
            'porcentaje_nulos': nulos_por_columna.sort_values(ascending=False).head(10)
        }))

    # Final distribution of the target variable
    print(f"\nüéØ FINAL TARGET VARIABLE DISTRIBUTION:")
    conteo_final = dataset[target_col].value_counts()
    for valor, conteo in conteo_final.items():
        etiqueta = {1: "Diabetes", 3: "No Diabetes", 4: "Prediabetes"}.get(valor, valor)
        porcentaje = conteo / len(dataset) * 100
        print(f"  ‚Ä¢ {etiqueta}: {conteo:,} ({porcentaje:.1f}%)")

    # Calculate class balance
    balance = conteo_final.max() / conteo_final.min()
    print(f"  ‚Ä¢ Balance (largest/smallest class): {balance:.2f}:1")

    # Numeric vs Categorical variables
    numeric_cols = dataset.select_dtypes(include=np.number).columns
    categorical_cols = dataset.select_dtypes(include=['category', 'object']).columns

    print(f"\nüî¢ NUMERIC VARIABLES: {len(numeric_cols)}")
    print(f"üìù CATEGORICAL VARIABLES: {len(categorical_cols)}")

    return {
        'dimensiones': dataset.shape,
        'memoria_mb': dataset.memory_usage(deep=True).sum() / 1024**2,
        'nulos_totales': nulos_totales,
        'balance_clases': balance,
        'columnas_numericas': len(numeric_cols),
        'columnas_categoricas': len(categorical_cols)
    }

metricas_finales = analisis_final_dataset_limpio(dataset_limpio)

# ------------------------------------
# 9. RECOMMENDATIONS FOR NEXT STEPS
# ------------------------------------
print("\n" + "="*80)
print("üöÄ RECOMMENDATIONS FOR NEXT STEPS")
print("="*80)

print("""
1. üéØ **PROBLEM DEFINITION:**
   ‚Ä¢ Binary classification (Diabetes vs No Diabetes)?
   ‚Ä¢ Multiclass classification (Diabetes vs Prediabetes vs No Diabetes)?
   ‚Ä¢ Risk detection (Diabetes/Prediabetes vs Healthy)?

2. üîß **ADDITIONAL PREPROCESSING:**
   ‚Ä¢ Handling missing values (specific imputation per variable)
   ‚Ä¢ Encoding categorical variables
   ‚Ä¢ Scaling numerical variables
   ‚Ä¢ Class balancing if necessary

3. üìä **CORRELATION ANALYSIS:**
   ‚Ä¢ Identify highly correlated variables
   ‚Ä¢ Analyze relationship with target variable
   ‚Ä¢ Dimensionality reduction if necessary

4. ü§ñ **MODELING:**
   ‚Ä¢ Baseline with a simple model (Logistic Regression)
   ‚Ä¢ Complex models (Random Forest, XGBoost, Neural Networks)
   ‚Ä¢ Robust cross-validation
   ‚Ä¢ Feature importance analysis

5. üìà **EVALUATION:**
   ‚Ä¢ Appropriate metrics for imbalanced dataset
   ‚Ä¢ ROC curves, confusion matrix
   ‚Ä¢ Model interpretability
""")

print(f"\nüí° **KEY STATISTICS FOR DECISIONS:**")
print(f"   ‚Ä¢ Useful dataset size: {metricas_finales['dimensiones'][0]:,} records")
print(f"   ‚Ä¢ Class balance: {metricas_finales['balance_clases']:.2f}:1")
print(f"   ‚Ä¢ Available variables: {metricas_finales['dimensiones'][1]}")
print(f"   ‚Ä¢ Remaining missing values: {metricas_finales['nulos_totales']:,}")

print("\nüéâ **EXPLORATORY ANALYSIS SUCCESSFULLY COMPLETED**")

# ------------------------------------
# 10. SAVING THE CLEAN DATASET
# ------------------------------------
def guardar_dataset_limpio(dataset, ruta_original):
    """
    Saves the clean dataset for future use
    """
    ruta_guardado = ruta_original.replace('.csv', '_limpio.csv')
    dataset.to_csv(ruta_guardado, index=False)
    print(f"\nüíæ Clean dataset saved to: {ruta_guardado}")
    return ruta_guardado

ruta_limpio = guardar_dataset_limpio(dataset_limpio, ruta)

print("\n" + "="*80)
print("‚úÖ PROCESS COMPLETED - READY FOR THE NEXT PHASE")
print("="*80)
```
