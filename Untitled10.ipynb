```python
# =============================================================================
# AN√ÅLISIS EXPLORATORIO AVANZADO - DATASET BRFSS DIABETES
# =============================================================================

# ------------------------------------
# 1. CONFIGURACI√ìN PROFESIONAL COMPLETA
# ------------------------------------
"""
CONFIGURACI√ìN INTEGRAL: Elimina repeticiones y consolida todas las librer√≠as
y configuraciones en un solo bloque organizado.
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from scipy import stats
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
import missingno as msno
from scipy.cluster import hierarchy
from scipy.stats import pointbiserialr
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Configuraci√≥n unificada para todo el an√°lisis
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 30)
np.random.seed(42)

print("‚úÖ Entorno profesional configurado - Listo para an√°lisis avanzado")

# ------------------------------------
# 2. CARGA OPTIMIZADA DEL DATASET
# ------------------------------------
def cargar_dataset_optimizado(ruta):
    """
    CARGA INTELIGENTE: Combina optimizaci√≥n de memoria con an√°lisis de calidad
    en una sola funci√≥n sin repeticiones.
    """
    print("üöÄ CARGANDO DATASET CON OPTIMIZACI√ìN AVANZADA...")

    # Muestra para inferencia de tipos
    muestra = pd.read_csv(ruta, encoding='latin-1', nrows=10000)

    # Optimizaci√≥n de tipos de datos
    dtype_optimizado = {}
    for col in muestra.columns:
        col_data = muestra[col]

        if col_data.dtype == 'object':
            unique_ratio = col_data.nunique() / len(col_data)
            dtype_optimizado[col] = 'category' if unique_ratio < 0.5 else 'object'

        elif pd.api.types.is_integer_dtype(col_data):
            c_min, c_max = col_data.min(), col_data.max()
            if c_min >= 0:
                dtype_optimizado[col] = 'uint8' if c_max < 256 else 'uint16' if c_max < 65535 else 'uint32'
            else:
                dtype_optimizado[col] = 'int16' if c_min > -32768 and c_max < 32767 else 'int32'
        else:
            dtype_optimizado[col] = 'float32'

    # Carga completa optimizada
    df = pd.read_csv(ruta, encoding='latin-1', dtype=dtype_optimizado, low_memory=False)

    # An√°lisis de memoria
    memoria_mb = df.memory_usage(deep=True).sum() / 1024**2
    print(f"‚úÖ Dataset cargado: {df.shape[0]:,} filas √ó {df.shape[1]} columnas")
    print(f"üíæ Memoria utilizada: {memoria_mb:.2f} MB")

    return df

# Carga del dataset
try:
    from google.colab import drive
    drive.mount('/content/drive')
    ruta = '/content/drive/MyDrive/Copia de Dataset_sin_tratar.csv'
except:
    ruta = 'Copia de Dataset_sin_tratar.csv'

dataset = cargar_dataset_optimizado(ruta)

# ------------------------------------
# 3. AN√ÅLISIS EXPLORATORIO INTEGRADO
# ------------------------------------
def analisis_exploratorio_integrado(df):
    """
    AN√ÅLISIS CONSOLIDADO: Combina an√°lisis estructural, de missing values,
    target y cardinalidad en un flujo coherente.
    """
    print("\n" + "="*80)
    print("üîç AN√ÅLISIS EXPLORATORIO INTEGRADO")
    print("="*80)

    # 3.1 An√°lisis Estructural
    print(f"üìê DIMENSIONES: {df.shape[0]:,} filas √ó {df.shape[1]} columnas")

    tipo_dist = df.dtypes.value_counts()
    print("\nüìä DISTRIBUCI√ìN DE TIPOS DE DATOS:")
    for tipo, count in tipo_dist.items():
        porcentaje = count / df.shape[1] * 100
        print(f"   ‚Ä¢ {str(tipo):<15} {count:>3} columnas ({porcentaje:5.1f}%)")

    # Quick Descriptive Statistics
    print("\nüìã DESCRIPTIVE STATISTICS (first 10 numeric columns):")
    numeric_cols = df.select_dtypes(include=np.number).columns[:10]
    display(df[numeric_cols].describe())

    # 3.2 An√°lisis de Missing Values Avanzado
    print("\n" + "-"*50)
    print("üîç AN√ÅLISIS DE VALORES FALTANTES")
    print("-"*50)

    # An√°lisis cuantitativo
    nulos_por_columna = df.isnull().sum()
    porcentaje_nulos = (nulos_por_columna / len(df)) * 100

    categorias_missing = {
        'Cr√≠tico (>70%)': porcentaje_nulos[porcentaje_nulos > 70],
        'Alto (30-70%)': porcentaje_nulos[(porcentaje_nulos > 30) & (porcentaje_nulos <= 70)],
        'Moderado (5-30%)': porcentaje_nulos[(porcentaje_nulos > 5) & (porcentaje_nulos <= 30)],
        'Bajo (<5%)': porcentaje_nulos[porcentaje_nulos <= 5]
    }

    print("üìä CLASIFICACI√ìN DE MISSING VALUES:")
    for categoria, series in categorias_missing.items():
        print(f"   ‚Ä¢ {categoria}: {len(series)} variables")


    # 3.3 An√°lisis de Variable Target (DIABETE3)
    print("\n" + "-"*50)
    print("üéØ AN√ÅLISIS DE VARIABLE TARGET")
    print("-"*50)

    if 'DIABETE3' in df.columns:
        codigos_diabetes = {
            1: "Diabetes", 2: "Diabetes Gestacional", 4: "Prediabetes",
            3: "No Diabetes", 7: "No sabe/No seguro", 9: "Se neg√≥ a responder"
        }

        distribucion = df['DIABETE3'].value_counts().sort_index()

        # Visualizaci√≥n integrada
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

        # Gr√°fico de barras
        bars = ax1.bar([codigos_diabetes.get(x, f'Valor {x}') for x in distribucion.index],
                      distribucion.values, color=['red', 'pink', 'green', 'lightgreen', 'gray', 'darkgray'])
        ax1.set_title('Distribuci√≥n de DIABETE3', fontweight='bold')
        ax1.tick_params(axis='x', rotation=45)

        # A√±adir etiquetas
        for bar in bars:
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:,}\n({height/len(df)*100:.1f}%)',
                    ha='center', va='bottom', fontsize=8)

        # Pie chart de clases principales
        clases_principales = {k: v for k, v in codigos_diabetes.items() if k in [1, 3, 4]}
        counts_principales = [distribucion.get(k, 0) for k in clases_principales.keys()]
        ax2.pie(counts_principales, labels=clases_principales.values(), autopct='%1.1f%%', startangle=90)
        ax2.set_title('Clases Principales para Modelado')

        # An√°lisis de balance
        clases_modelado = {'Diabetes': distribucion.get(1, 0), 'No Diabetes': distribucion.get(3, 0)}
        ax3.bar(clases_modelado.keys(), clases_modelado.values(), color=['red', 'green'])
        ax3.set_title('Balance para Clasificaci√≥n Binaria')

        # Distribuci√≥n acumulativa
        distribucion_acum = distribucion.cumsum() / distribucion.sum()
        ax4.plot(range(len(distribucion_acum)), distribucion_acum.values, 'o-')
        ax4.set_title('Distribuci√≥n Acumulativa')
        ax4.set_xlabel('Categor√≠as')
        ax4.set_ylabel('Proporci√≥n Acumulada')

        plt.tight_layout()
        plt.show()

        # Estad√≠sticas del target
        total_util = sum(distribucion.get(k, 0) for k in [1, 3, 4])
        print(f"üìà ESTAD√çSTICAS TARGET:")
        print(f"   ‚Ä¢ Registros √∫tiles para modelado: {total_util:,} ({total_util/len(df)*100:.1f}%)")
        print(f"   ‚Ä¢ Ratio de balance: {max(distribucion)/min(distribucion):.1f}:1")

    # 3.4 An√°lisis de Cardinalidad Integrado
    print("\n" + "-"*50)
    print("üìä AN√ÅLISIS DE CARDINALIDAD")
    print("-"*50)

    resumen_cardinalidad = []
    for col in df.columns:
        n_unique = df[col].nunique()
        resumen_cardinalidad.append({
            'columna': col,
            'valores_unicos': n_unique,
            'porcentaje_unicos': (n_unique / len(df)) * 100,
            'es_constante': n_unique == 1,
            'tipo_dato': df[col].dtype
        })

    card_df = pd.DataFrame(resumen_cardinalidad)

    # Resumen de problemas de cardinalidad
    constantes = card_df[card_df['es_constante']]
    alta_cardinalidad = card_df[card_df['porcentaje_unicos'] > 90]

    print("üö® PROBLEMAS IDENTIFICADOS:")
    print(f"   ‚Ä¢ Variables constantes: {len(constantes)}")
    print(f"   ‚Ä¢ Variables alta cardinalidad: {len(alta_cardinalidad)}")

    if len(constantes) > 0:
        print(f"\nüìã VARIABLES CONSTANTES (eliminar):")
        for col in constantes['columna'].head(5):
            print(f"   ‚Ä¢ {col}")

    return df

# Carga del dataset
try:
    from google.colab import drive
    drive.mount('/content/drive')
    ruta = '/content/drive/MyDrive/Copia de Dataset_sin_tratar.csv'
except:
    ruta = 'Copia de Dataset_sin_tratar.csv'

dataset = cargar_dataset_optimizado(ruta)

# Ejecutar an√°lisis exploratorio integrado
dataset_original = dataset.copy() # Keep a copy of original dataset
dataset = analisis_exploratorio_integrado(dataset)

# ------------------------------------
# 4. MISSING VALUES ANALYSIS
# ------------------------------------
def analisis_missing_values(dataset, umbral_alto=70, umbral_medio=30):
    """
    Detailed analysis of missing values with treatment strategies
    """
    print("\n" + "="*80)
    print("üîç MISSING VALUES ANALYSIS")
    print("="*80)

    total_filas = len(dataset)
    nulos_por_columna = dataset.isnull().sum()
    porcentaje_nulos = (nulos_por_columna / total_filas) * 100

    resumen_nulos = pd.DataFrame({
        'valores_nulos': nulos_por_columna,
        'porcentaje_nulos': porcentaje_nulos,
        'tipo_dato': dataset.dtypes
    }).sort_values('porcentaje_nulos', ascending=False)

    # Classify columns by missing value level
    columnas_problematicas = resumen_nulos[resumen_nulos['porcentaje_nulos'] > umbral_alto]
    columnas_manejables = resumen_nulos[
        (resumen_nulos['porcentaje_nulos'] > umbral_medio) &
        (resumen_nulos['porcentaje_nulos'] <= umbral_alto)
    ]
    columnas_limpias = resumen_nulos[resumen_nulos['porcentaje_nulos'] <= umbral_medio]

    print(f"\nüìä MISSING VALUES SUMMARY:")
    print(f"‚Ä¢ Columns with >{umbral_alto}% nulls: {len(columnas_problematicas)}")
    print(f"‚Ä¢ Columns with {umbral_medio}-{umbral_alto}% nulls: {len(columnas_manejables)}")
    print(f"‚Ä¢ Columns with <{umbral_medio}% nulls: {len(columnas_limpias)}")
    print(f"‚Ä¢ Total null values in dataset: {dataset.isnull().sum().sum():,}")

    # Show the top 10 most problematic columns
    if len(columnas_problematicas) > 0:
        print(f"\nüö® TOP 10 MOST PROBLEMATIC COLUMNS (> {umbral_alto}% nulls):")
        display(columnas_problematicas.head(10))

    return resumen_nulos, columnas_problematicas, columnas_manejables, columnas_limpias

resumen_nulos, problematicas, manejables, limpias = analisis_missing_values(dataset_original)

# ------------------------------------
# 5. TARGET VARIABLE ANALYSIS (DIABETE3)
# ------------------------------------
def analisis_variable_target(dataset, target_col='DIABETE3'):
    """
    Exhaustive analysis of the target variable for classification problems
    """
    print("\n" + "="*80)
    print("üéØ TARGET VARIABLE ANALYSIS: DIABETE3")
    print("="*80)

    # BRFSS code dictionary (according to official documentation)
    codigos_diabetes = {
        1: "S√≠, diabetes",
        2: "S√≠, but only during pregnancy",
        3: "No, does not have diabetes",
        4: "S√≠, prediabetes or borderline",
        7: "Does not know / Not sure",
        9: "Refused to answer"
    }

    # Detailed distribution
    conteos = dataset[target_col].value_counts().sort_index()
    porcentajes = (conteos / len(dataset)) * 100

    print("\nüìä TARGET VARIABLE DISTRIBUTION:")
    for valor, conteo in conteos.items():
        etiqueta = codigos_diabetes.get(valor, f"Valor {valor}")
        print(f"  {valor}: {etiqueta:.<40} {conteo:>8,} ({porcentajes[valor]:.2f}%)")

    # Analysis of useful classes for modeling
    clases_principales = {1: "Diabetes", 3: "No Diabetes", 4: "Prediabetes"}
    clases_excluir = {2: "Gestational", 7: "Does not know", 9: "Refused"}

    total_util = sum(conteos.get(k, 0) for k in clases_principales.keys())
    total_excluir = sum(conteos.get(k, 0) for k in clases_excluir.keys())

    print(f"\nüéØ CLASSES FOR MODELING:")
    for valor, etiqueta in clases_principales.items():
        if valor in conteos:
            count = conteos[valor]
            print(f"  ‚Ä¢ {etiqueta}: {count:,} records ({count/len(dataset)*100:.2f}%)")

    print(f"\nüö´ CLASSES TO EXCLUDE:")
    for valor, etiqueta in clases_excluir.items():
        if valor in conteos:
            count = conteos[valor]
            print(f"  ‚Ä¢ {etiqueta}: {count:,} records ({count/len(dataset)*100:.2f}%)")

    print(f"\nüìà CLASS BALANCE:")
    print(f"  ‚Ä¢ Useful records for modeling: {total_util:,} ({total_util/len(dataset)*100:.2f}%)")
    print(f"  ‚Ä¢ Records to exclude: {total_excluir:,} ({total_excluir/len(dataset)*100:.2f}%)")

    if total_util > 0:
        for valor, etiqueta in clases_principales.items():
            if valor in conteos:
                proporcion = conteos[valor] / total_util * 100
                print(f"  ‚Ä¢ {etiqueta} in useful dataset: {conteos[valor]:,} ({proporcion:.1f}%)")

    # Visualization of the distribution
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    datos_plot = dataset[target_col].value_counts().sort_index()
    bars = plt.bar([codigos_diabetes.get(x, x) for x in datos_plot.index], datos_plot.values)
    plt.title('Distribution of DIABETE3', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Frequency')

    # Add labels to bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:,}\n({height/len(dataset)*100:.1f}%)',
                ha='center', va='bottom', fontsize=8)

    plt.subplot(1, 2, 2)
    # Only main classes for pie chart
    clases_para_pie = {k: v for k, v in clases_principales.items() if k in conteos}
    sizes = [conteos[k] for k in clases_para_pie.keys()]
    labels = [f'{v}\n({sizes[i]/total_util*100:.1f}%)' for i, v in enumerate(clases_para_pie.values())]

    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')
    plt.title('Distribution of Main Classes\n(for modeling)', fontsize=14, fontweight='bold')

    plt.tight_layout()
    plt.show()

    return clases_principales, total_util

clases_principales, total_util = analisis_variable_target(dataset_original)

# ------------------------------------
# 6. CARDINALITY AND CONSTANT VARIABLES ANALYSIS
# ------------------------------------
def analisis_cardinalidad_constantes(dataset):
    """
    Identifies variables with high cardinality and constant variables
    """
    print("\n" + "="*80)
    print("üìä CARDINALITY AND CONSTANT VARIABLES ANALYSIS")
    print("="*80)

    total_filas = len(dataset)
    resumen_cardinalidad = []

    for columna in dataset.columns:
        valores_unicos = dataset[columna].nunique()
        porcentaje_unicos = (valores_unicos / total_filas) * 100
        es_constante = valores_unicos == 1
        tipo_dato = dataset[columna].dtype

        resumen_cardinalidad.append({
            'columna': columna,
            'valores_unicos': valores_unicos,
            'porcentaje_unicos': porcentaje_unicos,
            'es_constante': es_constante,
            'tipo_dato': tipo_dato
        })

    df_cardinalidad = pd.DataFrame(resumen_cardinalidad)

    # Constant variables
    variables_constantes = df_cardinalidad[df_cardinalidad['es_constante']]
    print(f"\nüö® CONSTANT VARIABLES (recommended to drop): {len(variables_constantes)}")
    if len(variables_constantes) > 0:
        display(variables_constantes[['columna', 'valores_unicos', 'tipo_dato']])

    # Variables with high cardinality (>90% unique values)
    alta_cardinalidad = df_cardinalidad[df_cardinalidad['porcentaje_unicos'] > 90]
    print(f"\n‚ö†Ô∏è VARIABLES WITH HIGH CARDINALITY (>90% unique): {len(alta_cardinalidad)}")
    if len(alta_cardinalidad) > 0:
        display(alta_cardinalidad[['columna', 'valores_unicos', 'porcentaje_unicos', 'tipo_dato']].head(10))

    # Variables with low cardinality (<5% unique values - potential categorical)
    baja_cardinalidad = df_cardinalidad[
        (df_cardinalidad['porcentaje_unicos'] < 5) &
        (~df_cardinalidad['es_constante']) &
        (df_cardinalidad['valores_unicos'] > 1)
    ]
    print(f"\nüéØ VARIABLES WITH LOW CARDINALITY (<5% unique): {len(baja_cardinalidad)}")
    if len(baja_cardinalidad) > 0:
        display(baja_cardinalidad[['columna', 'valores_unicos', 'porcentaje_unicos', 'tipo_dato']].head(10))

    return df_cardinalidad, variables_constantes, alta_cardinalidad, baja_cardinalidad

df_cardinalidad, constantes, alta_card, baja_card = analisis_cardinalidad_constantes(dataset_original)
```
